```python
df['granthamScore'] = pd.to_numeric(df['granthamScore'], errors='coerce')
df['scoreCADD'] = pd.to_numeric(df['scoreCADD'], errors='coerce')
df['polyPhen'] = pd.to_numeric(df['polyPhen'], errors='coerce')

high_impact = df[(df['functionGVS'] == 'missense') & 
                       (df['granthamScore'] > 100) &
                          (df['scoreCADD'] > 20) &
                            (df['polyPhen'] > 0.85)
                       ]

priority_variants = pd.concat([high_impact]).drop_duplicates()
priority_variants
```

```python
priority_variants.to_excel('priority_variants.xlsx', index=False)
```
